1. Write a python program to implement K-means Clustering algorithm

o Generate 10000 2-D data points in the range 0-100 randomly
o Divide data points into 5 clusters
o Find time taken by the algorithm to find clusters
2. Write a python program to implement Mini-batch K-means Clustering algorithm
o Generate 10000 2-D data points in the range 0-100 randomly
o Divide data points into 5 clusters
o Find time taken by the algorithm to find clusters
o Vary the batch size from 100 to 1500, find time taken by the algorithm in
each case and find best value of the batch size.

```python
# 1 number
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = np.random.rand(10000, 2) * 100

km = KMeans(n_clusters=5, init="random")

t0 = time.process_time()
km.fit(data)
t1 = time.process_time()

tt = t1 - t0
print("Total Time:", tt)

centers = km.cluster_centers_
labels = km.labels_

print("Cluster Centers:", centers)
print("Cluster Labels:", *labels)


```

```python
colors = ["g", "r", "b", "y", "m"]
markers = ["+", "x", "*", ".", "d"]

for i in range(len(data)):
    plt.plot(data[i][0], data[i][1], color=colors[labels[i]], marker=markers[labels[i]])
plt.scatter(centers[:, 0], centers[:, 1], marker="o", s=50, linewidths=5)
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans

# Generate 10,000 2-D data points in the range 0-100 randomly
data = np.random.rand(10000, 2) * 100

# Function to find the best batch size
def find_best_batch_size(data, num_clusters=5, batch_size_range=(100, 1500), num_iterations=10):
    best_batch_size = None
    best_time = float('inf')
    for batch_size in range(batch_size_range[0], batch_size_range[1] + 1):
        mbk = MiniBatchKMeans(n_clusters=num_clusters, init="random", batch_size=batch_size)
        t0 = time.time()
        mbk.fit(data)
        t1 = time.time()
        elapsed_time = t1 - t0
        print(f"Batch size: {batch_size}, Time taken: {elapsed_time:.2f} seconds")
        if elapsed_time < best_time:
            best_time = elapsed_time
            best_batch_size = batch_size
    return best_batch_size, best_time

# Find the best batch size
best_batch_size, best_time = find_best_batch_size(data)

print(f"Best batch size: {best_batch_size}, Best time: {best_time:.2f} seconds")

# Use the best batch size to fit the data
mbk = MiniBatchKMeans(n_clusters=5, init="random", batch_size=best_batch_size)
t0 = time.time()
mbk.fit(data)
t1 = time.time()

tt = t1 - t0
print("Total Time: ", tt)

centers = mbk.cluster_centers_
labels = mbk.labels_

print("Cluster Centers:", centers)
print("Labels:", labels)

colors = ["g", "r", "b", "y", "m"]
markers = ["+", "x", "*", ".", "d"]

for i in range(len(data)):
    plt.plot(data[i][0], data[i][1], color=colors[labels[i]], marker=markers[labels[i]])
plt.scatter(centers[:, 0], centers[:, 1], marker="o", s=50, linewidths=5)
plt.show()
```
